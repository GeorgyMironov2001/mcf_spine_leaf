```python
def read_tests(name, dir_name):
    res_tests = {}
    directory_path = f'{dir_name}/{name}'
    for alpha in os.listdir(directory_path):
        full_path = os.path.join(directory_path, alpha)
        count = int(len([entry for entry in os.listdir(full_path) if os.path.isfile(os.path.join(full_path, entry))])/2)
        for num in range(count):
            test_data = None
            test_color = None
            with open(os.path.join(full_path, f"test_{num}.json"), 'r') as f:
                test_data = json.load(f)
            with open(os.path.join(full_path, f"color_{num}.json"), 'r') as f:
                test_color = json.load(f)
            if alpha not in res_tests:
                res_tests[alpha] = []
            res_tests[alpha].append([[test_data[i], test_color[i]] for i in range(len(test_color))])
    return res_tests
def read_results(name, dir_name):
    result_file = os.path.join(dir_name, f'{name}.json')
    res = None
    with open(result_file, 'r') as f:
        res = json.load(f)
    return res
```

```python
def split_results(tests, results):
    new_results = {}
    for alpha in alpha_list:
        ptr = 0
        new_results[alpha] = []
        for test in tests[alpha]:
            new_results[alpha].append(results[alpha][ptr:ptr+len(test)])
            ptr += len(test)
    return new_results
```

```python
%%time
dir_tests = {
    "multi_tasks_tests": ['2_32_16', '5_16_8', '3_8_4', '4_8_4', '3_16_8', '8_16_8', '5_8_4', '4_16_8', '6_16_8', '7_16_8', '2_16_8', '2_8_4'],
    "new_multi_tasks_tests": ['3_16_4', '5_16_8', '7_16_4', '8_16_4', '4_16_4', '8_16_8', '4_16_8', '6_16_4', '2_16_4', '5_16_4', '6_16_8', '7_16_8', '2_32_8']
}
all_tests = {}
results1 = {}
results2 = {}
for test_dir, tls in dir_tests.items():
    for t_l_s in tls:
        if t_l_s not in all_tests:
            all_tests[t_l_s], results1[t_l_s], results2[t_l_s] = {}, {}, {}
        tls_tests = read_tests(t_l_s, test_dir)
        res1 = split_results(tls_tests, read_results(t_l_s, f'{test_dir}_results_scenario1'))
        res2 = split_results(tls_tests, read_results(t_l_s, f'{test_dir}_results_scenario2'))

        for alpha in alpha_list:
            for i, test_list in enumerate(tls_tests[alpha]):
                distr = tuple([sum(x) for x in test_list[0][0]])
                if sum(distr) not in all_tests[t_l_s]:
                    all_tests[t_l_s][sum(distr)] = {}
                    results1[t_l_s][sum(distr)] = {}
                    results2[t_l_s][sum(distr)] = {}
                if distr not in all_tests[t_l_s][sum(distr)]:
                    all_tests[t_l_s][sum(distr)][distr] = {alpha_:[] for alpha_ in alpha_list}
                    results1[t_l_s][sum(distr)][distr] = {alpha_:[] for alpha_ in alpha_list}
                    results2[t_l_s][sum(distr)][distr] = {alpha_:[] for alpha_ in alpha_list}
                for j, test in enumerate(test_list):
                    all_tests[t_l_s][sum(distr)][distr][alpha].append(test)
                    results1[t_l_s][sum(distr)][distr][alpha].append(res1[alpha][i][j])
                    results2[t_l_s][sum(distr)][distr][alpha].append(res2[alpha][i][j])
```

```python
def make_plot(tls, distr_sum, all_tests, results1, results2):
    # fig, axs = plt.subplots(figsize=(10, 8))
    stats_1 = {alpha: [0, 0] for alpha in alpha_list}
    stats_2 = {alpha: [0, 0] for alpha in alpha_list}
    
    for distr, alpha_tests in all_tests[tls][distr_sum].items():
        for alpha in alpha_list:
            for i, test in enumerate(alpha_tests[alpha]):
                if not test[1]:
                    continue
                stats_1[alpha][results1[tls][distr_sum][distr][alpha][i]] += 1
                stats_2[alpha][results2[tls][distr_sum][distr][alpha][i]] += 1

    return stats_1 ,stats_2
```
